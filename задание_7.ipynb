```python
#7.5
while True:
    v=input("Введите выражение, отделяя числа и знаки действия пробелами(чтобы выйти введите 'exit'):")
    if v=="exit":
        print("Завершение работы! Хорошего дня!")
        break
    x, de, y=part=v.split()
    x=float(x)
    y=float(y)
    match de:
        case "+":
            print(x+y)
        case "-":
            print(x-y)
        case "/":
            if y!=0:
                print(x/y)
            else:
                print("Действие невозможно!")
        case "*":
            print(x*y)
        
```

```python
#7.7
l=input("Введите интервал(начало и конец через пробел):")
a, z=l.split()
a=int(a)
z=int(z)
for o in range(a,z+1):
    if o<2:
        continue
    p=True
    for d in range(2, o):
        if o%d==0:
            p=False
            break
    if p:
        print(o)
```

```python
#7.8
b=input("Введите значение Начальной суммы, Годового процента и Конечной суммы через пробел:")
x, p, y=b.split()
k=0
x=int(x)
p=int(p)
y=int(y)
while y>x:
    x=x + (x*p/100)
    x=int(x)
    k+=1
    if x>=y:
        break
print(k)
```

```python
#7.9
g=input("Введите числа через пробел, в конце 0: ")
part=g.split()
y=int(part[0])
max1=y
max2=None
for i in range(1,len(part)):
    y=int(part[i])
    if y==0:
        break
    if max2 is None:
        if y>max1:
            max2=max1
            max1=y
        else:
            max2=y
    else:
        if y>max1:
            max2, max1=max1, y
        elif y>max2:
            max2=y
print(g)
print(max2)
```
