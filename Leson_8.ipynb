```python
#4.1.4
f=[]
while True:
    d=input('Введите число:')
    k=int(d)
    if k==0:
        break
    f.append(k)
minus=False        #наличие -
for i in f:
    if i<0:
        minus=True
        break
if minus:
    for j in range(len(f)):
        f[j]= -f[j]
print(f)
```

```python
#4.1.5
a=[1,2,3,4,5,6]
ch=0
nch=0
for i,v in enumerate(a):        #превращает список в последовательность пар(позиция и значение)
    if i%2==0:
        ch+=v
    else:
        nch+=v
print(ch/nch)
```

```python
#4.2
a1=['A', 'E', 'I', 'O', 'U', 'L', 'N', 'S', 'T', 'R']
a2=['D', 'G']
a3=['B', 'C', 'M', 'P']
a4=['F', 'H', 'V', 'W', 'Y']
a5=['K']
a8=['J', 'X']
a10=['Q', 'Z']
r1=['А', 'В', 'Е', 'И', 'Н', 'О', 'Р', 'С', 'Т']
r2=['Д', 'К', 'Л', 'М', 'П', 'У']
r3=['Б', 'Г', 'Ё', 'Ь', 'Я']
r4=['Й', 'Ы']
r5=['Ж', 'З', 'Х', 'Ц', 'Ч']
r8=['Ш', 'Э', 'Ю']
r10=['Ф', 'Щ', 'Ъ']
k=0
r=input('Введите слова заглавными буквами:')
for i in r:
    if i in a1 or i in r1:
        k+=1
    if i in a2 or i in r2:
        k+=2
    if i in a3 or i in r3:
        k+=3
    if i in a4 or i in r4:
        k+=4
    if i in a5 or i in r5:
        k+=5
    if i in a8 or i in r8:
        k+=8
    if i in a10 or i in r10:
        k+=10
print(r,k)
```

```python
#4.3
cats = [
    ("Муся", 7, "Игорь", "Бероев"),
    ("Барсик", 5, "Игорь", "Бероев"),
    ("Изольда", 2, "Игорь", "Бероев"),
    ("Рыжик", 3, "Анна", "Петрова"),
    ("Снежок", 1, "Анна", "Петрова")
]
d=list(cats)
s={}        #словар
print(d)
for cat,age,name,surname in cats:
    o=name+' '+surname
    if o not in s:
        s[o]=[]     #словарь s, в нем запись с ключом o и туда пустой список [] в качестве значения
    s[o].append(f"{cat}, {age}")
for owner, cats in s.items():        #items() превращает словарь в список пар "ключ -> значение"
    print(f"{owner}: {'; '.join(cats)}")
```

```python
#4.5
import random
tower1=[]
tower2=[]
tower3=[]
rings=[1,2,3,4,5,6,7]
random.shuffle(rings)       #рандомно перемешивает
tower1=rings
moves =0
print("Башня1:", tower1)
print("Башня2:", tower2)
print("Башня3:", tower3)
print("Цель: собрать [7,6,5,4,3,2,1] на любой башне")

while True:
    if tower1==[7,6,5,4,3,2,1] or tower2==[7,6,5,4,3,2,1] or tower3==[7,6,5,4,3,2,1]:
        print("ПОБЕДА! Ходов:", moves)
        break
    move_tower=input("Введите с какой на какую башну хотите переложть (1-3, через пробел):")
    if move_tower=="0":
        break
    a,b=move_tower.split()
    from_tower=int(a)
    to_tower=int(b)
    
    if from_tower==1:
        list_from=tower1
    elif from_tower==2:
        list_from=tower2
    else:
        list_from=tower3
        
    if to_tower==1:
        list_to=tower1
    elif to_tower==2:
        list_to=tower2
    else:
        list_to=tower3

    if len(list_from)==0:
        print("Пусто")
    elif len(list_to)>0 and list_from[-1] > list_to[-1]:
        print("Нельзя большее на меньшее:")
    else:
        ring=list_from.pop()        #Удаляет последний элемент из списка и возвращает его
        list_to.append(ring)
        moves+=1

        print("Башня1:", tower1)
        print("Башня2:", tower2)
        print("Башня3:", tower3)
        print("Ходов:", moves)
print("Конец!")
```

```python
#4.4
import random
import time

N=10
numbers=[]
for m in range(N):
    numbers.append(random.randint(1,100))       #генерируем случайное число от 1 до 100
print("Исходный массив:", numbers)

sorted_numbers=numbers.copy()
n=len(sorted_numbers)
for i in range(n-1):
    for j in range(n-1-i):
        if sorted_numbers[j]>sorted_numbers[j+1]:
            sorted_numbers[j], sorted_numbers[j+1]=sorted_numbers[j+1], sorted_numbers[j]
print("Отсортированный:", sorted_numbers)

print("="*55)

for size in[10,100,1000,10000]:
    q=[]
    for w in range(size):
        q.append(random.randint(1,100))
    start=time.time()
    sorted_q=q.copy()
    n=len(sorted_q)
    for i in range(n-1):
        for j in range(n-1-i):
            if sorted_q[j]>sorted_q[j+1]:
                sorted_q[j], sorted_q[j+1]=sorted_q[j+1],sorted_q[j]

    end=time.time()
    print(f"{size:<10} {end-start:15.6f} сек")
```
